### **Task 3**: Missing Values Handling

Firstly, The Columns Notes and Status are competely empty where have missing values with all the rows of the dataframe, so the suitable method with these column is just drop them from the dataframe

```python
df.drop(columns=['Notes', 'Status'], inplace=True)
df.head()
```

Let's see the Benefits Column

```python
import seaborn as sns

sns.histplot(data=df['Benefits'].fillna(-10000))
```

We Can Impute Missing Values with Median Strategy using Pipeline Class, Where The Mean Stretegy can cause Baised Imputaion due to the larg proportion of the missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
df['Benefits'] = imputer.fit_transform(df[['Benefits']])
```

```python
sns.histplot(data=df, x='Benefits')
```

And Now Let's see BasePay Column

```python
sns.histplot(data=df.BasePay.fillna(-10000))
```

### **Task 4**: Pie Chart

```python
import matplotlib.pyplot as plt

# Extract department names from job titles using regular expressions and convert to lowercase
df['Department'] = df['JobTitle'].apply(lambda x: x.split('(')[-1].split(')')[0].lower() if '(' in x else np.nan)

# Count the number of employees in each department
department_counts = df['Department'].value_counts()

# Create a pie chart using seaborn
plt.figure(figsize=(8, 6))
sns.set(style="darkgrid")
sns.set_palette("Set3")
plt.title("Proportion of Employees in Different Departments")
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.show()
```

### *There Are A lot of Missing Department in JobTitle So I will not plot them as unknown department, and the deference in upper and lower letters are considered*

### **Task 5**: Calculate The Avg Salary within Each Department

```python
# Calculate the average total pay within each department
avg_total_pay = df.groupby('Department')['TotalPay'].mean()

# Print the average total pay for each department
print(avg_total_pay)
```

# Task 6

### **Task 6**: Correlation Between BasePay and TotalPay

```python
# Calculate the correlation between 'BasePay' and 'TotalPay'
correlation = df['BasePay'].corr(df['TotalPay'])

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['BasePay'], df['TotalPay'])
plt.xlabel('Base Pay')
plt.ylabel('Total Pay')
plt.title('Relationship between Base Pay and Total Pay')
plt.grid(True)
plt.show()

# Print the correlation coefficient
print(f"Correlation between Base Pay and Total Pay: {correlation}")
```

## *The Correlation Result Shows that there is a strong relationship between BasePay and TotalPay Features, so may it is useful to drop on of them where the multicollinearity can be avoided*
